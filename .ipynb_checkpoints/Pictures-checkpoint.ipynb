```python

fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 360)
v = np.linspace(0, np.pi, 180)

x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

def get_colors(inp, colormap, vmin=None, vmax=None):
    norm = plt.Normalize(vmin, vmax)
    return colormap(norm(inp))
#norm = plt.Normalize()
#colors=cm.jet(norm(T_dis_sel_set3))
T=lna(T_dis_bud_set3)-273.15
colors=get_colors(T,plt.cm.jet,vmin=np.min(T),vmax=np.max(T))
mymap=LinearSegmentedColormap.from_list('mycm',colors,N=180)
#cmap_name2 = 'cmap2'
colors2 = [(1,0,0),(1,1,1),(0,0,1)] # red, white, blue
#CreateOwnColormap(cmap_name2, colors)
plot=ax.plot_surface(x, y, z,linewidth=0,cmap=mymap, antialiased=False, rstride=1, cstride=1)
#cset = ax.contourf(X, Y, Z, cmap=cm.coolwarm)
#ax.clabel(cset,zdir=Z, fontsize=9, inline=1)
#ax.contour(X, Y, Z, 10, linewidth=3, cmap=cm.coolwarm, linestyles="solid")
#ax.contour(X, Y, Z, 10, linewidth=3, colors="k", linestyles="solid")
ax.set_axis_off()

#ax.set_zlim(-1.01, 1.01)
#ax.zaxis.set_major_locator(LinearLocator(18))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(plot, shrink=0.5, aspect=5)
plt.show()
```
